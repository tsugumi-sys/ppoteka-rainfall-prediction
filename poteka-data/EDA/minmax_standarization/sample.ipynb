```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

root_dir = "../../../data/"
csv_filepath = "2020/10/2020-10-12/8-0.csv"
param_dirs = ["rain_image", "humidity_image", "temp_image"]
```

```python
rain_df = pd.read_csv(os.path.join(root_dir, param_dirs[0], csv_filepath), index_col=0)
data = rain_df.values.ravel()
_, axes = plt.subplots(1, 2, figsize=(16, 8))
axes = axes.ravel()

# Normal plot
sns.histplot(data, ax=axes[0], stat="density", cbar=False)
axes[0].set_xlabel("Rainfall intensity (mm/h)")

# Standarized plot
standarized_data = (data - data.mean()) / data.std()
sns.histplot(standarized_data, ax=axes[1], stat="density", cbar=False)
axes[1].set_xlabel("Rainfall intensity (mm/h)")


rain_df.head()
plt.tight_layout()
plt.show()
plt.close()
```

```python
humid_df = pd.read_csv(os.path.join(root_dir, param_dirs[1], csv_filepath), index_col=0)
data = humid_df.values.ravel()
_, axes = plt.subplots(2, 2, figsize=(16, 8))
axes = axes.ravel()

# Normal plot
sns.histplot(data, ax=axes[0], stat="density", cbar=False)
axes[0].set_xlabel("Humidity (%)")

# Standarized plot
standarized_data = (data - data.mean()) / data.std()
sns.histplot(standarized_data, ax=axes[1], stat="density", cbar=False)
axes[1].set_xlabel("Humidity (%)")


rain_df.head()
plt.tight_layout()
plt.show()
plt.close()
```

```python
temp_df = pd.read_csv(os.path.join(root_dir, param_dirs[2], csv_filepath), index_col=0)
data = temp_df.values.ravel()
_, axes = plt.subplots(1, 2, figsize=(16, 8))
axes = axes.ravel()

# Normal plot
sns.histplot(data, ax=axes[0], stat="density", cbar=False)
axes[0].set_xlabel("Temperature (℃)")

# Standarized plot
standarized_data = (data - data.mean()) / data.std()
sns.histplot(standarized_data, ax=axes[1], stat="density", cbar=False)
axes[1].set_xlabel("Temperature (℃)")


rain_df.head()
plt.tight_layout()
plt.show()
plt.close()
```
